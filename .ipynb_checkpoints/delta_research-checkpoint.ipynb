```python
import time
import pickle
import numpy as np
import matplotlib.pyplot as plt
from importlib import reload
from IPython.display import clear_output

from nn import Net
from utils import *
from layer import Layer
from activations import *

import keras
from keras.models import Sequential
from keras.layers import Dense
from keras.optimizers import SGD
from keras.datasets import mnist
from keras.datasets import cifar10
```

```python
y_logs = np.load('./logs/y_logs.npy')

activation_logs = np.load('./logs/activation_logs.npy')
delta_error_logs = np.load('./logs/delta_error_logs.npy')

activation_logs.shape, delta_error_logs.shape, y_logs.shape
```

```python
def sigmoid(arr, temp):
    return 1/(1 + np.exp(-arr*temp)) 

inp = np.linspace(-1,1,2000)
sig = sigmoid(inp, 10)
```

```python
neuron = 47 # np.random.randint(0,50,1)
xbound = np.max(np.abs(activation_logs [:,:,neuron]))
ybound = np.max(np.abs(delta_error_logs[:,:,neuron]))

for i in range(0,len(y_logs),20):
    mask = y_logs[i] < 2
    col  = y_logs[i][mask]
    act = activation_logs[i,:,neuron][mask]
    delta = delta_error_logs[i,:,neuron][mask]

    clear_output(wait=True)
    plt.figure(figsize=(10,10))
    
    plt.scatter(act, delta, c=col, alpha=0.7)
    plt.plot(inp, sig)
    plt.plot(inp, sig-1)

    plt.xlabel('activations')
    plt.ylabel('delta_error')
    plt.title("Neuron no. %d"%neuron)
    plt.xlim(-xbound*1.1, xbound*1.1)
    plt.ylim(-ybound*1.1, ybound*1.1)
    plt.show()
```
